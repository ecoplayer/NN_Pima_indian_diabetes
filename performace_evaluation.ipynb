```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import warnings
from keras.models import Sequential
from keras.layers import Dense

warnings.filterwarnings("ignore")
```

#### automatic verification

```python
### load pima indian dataset
path="data/Pima Indians Diabetes Database.csv"
df=pd.read_csv(path)
df.head(5)
#split into input (X) and output (Y) variables
X=df.copy()
del X["Class variable (0 or 1)"]
y=df["Class variable (0 or 1)"]
```

```python
#create model
num_features = 8 
model = Sequential()

model.add(Dense(12, input_dim=num_features, activation="relu"))
model.add(Dense(8, activation ="relu"))
model.add(Dense(12, activation ="relu"))
model.add(Dense(8, activation ="relu"))
model.add(Dense(1, activation="sigmoid"))

#compile model
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

#fit the model
model.fit(X,y, validation_split=0.33, epochs= 150, batch_size=10)
```

#### Manual Verification







```python
### load pima indian dataset
path="data/Pima Indians Diabetes Database.csv"
df=pd.read_csv(path)
df.head(5)
#split into input (X) and output (Y) variables
X=df.copy()
del X["Class variable (0 or 1)"]
y=df["Class variable (0 or 1)"]
#train test split
X_train, X_test, y_train, y_test =train_test_split(X,y, test_size=0.33)
```

```python
#fit the model
model.fit(X_train,y_train, validation_data=(X_test,y_test), epochs= 150, batch_size=10)
```

#### Cross validation

```python
from sklearn.model_selection import StratifiedKFold
### load pima indian dataset
path="data/Pima Indians Diabetes Database.csv"
df=pd.read_csv(path)
df.head(5)
#split into input (X) and output (Y) variables
X=df.copy()
del X["Class variable (0 or 1)"]
X=X.values
y=df["Class variable (0 or 1)"].values
```

```python
#define 10-fold cross validation test harness
kfold=StratifiedKFold(n_splits=10,shuffle=True) #no seed needed
cvscores=[]
for train, test in kfold.split(X,y):
      
    model.fit(X[train], y[train], epochs=150, batch_size=10,verbose=0)
    
    scores=model.evaluate(X[test],y[test], verbose=0)
    print("%s: %.2f%%" % (model.metrics_names[1],scores[1]*100))
    cvscores.append(scores[1]*100)
print("%.2f%%" % np.mean(cvscores),"(+/- %.2f%%)" % np.std(cvscores))
```
